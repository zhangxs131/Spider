```python
import requests
import lxml.etree as etree
import datetime
import time
import json
import os

#设置参数

keywords=['ukraine','Russia','punish','politics','military','world']
start_time='2022-03-07 00:00:00'
end_time='2022-03-10 00:00:00'
max_pages=10


#动态数据地址
url='https://cse.google.com/cse/element/v1?rsz=filtered_cse&num=10&hl=en&source=gcsc&gss=.com&start={}&cselibv=45f4e5efab1258be&cx=006972344228181832854:w07k6emi2wk&q={}&safe=active&cse_tok=AJvRUv1sR4qGRtiliiMd0n0yPlIs:1646882038496&lr=lang_en&sort=date&exp=csqr,cc&rsToken=undefined&afsExperimentId=undefined&callback=google.search.cse.api14656'

headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) '
                        'Chrome/51.0.2704.63 Safari/537.36'}

sum=0
result={}
for keyword in keywords:
  print('____________________under keyowrd {} _______________________'.format(keyword))
  data=[]
  for i in range(max_pages):
    search_url=url.format(i*10,keyword)
    try:
      r = requests.get(search_url, headers = headers, timeout = 30)
      r.raise_for_status()
      r.encoding = r.apparent_encoding
      t='{'+('{').join(r.text.split('{')[1:])
      t=('}').join(t.split('}')[:-1])+'}'
      js=json.loads(t)
    except:
      print('error in geting pages')

    for i in js['results']:
      sample={}
      t=i['richSnippet']
      if 'newsarticle' in t:
        sample['title']=t['newsarticle']['headline']
        if 'url' not in t['metatags']:
          continue
        sample['url']=t['metatags']['url']
        sample['time']=t['newsarticle']['datemodified'].replace('T',' ')[:-1]
        sample['text']=t['newsarticle']['articlebody']
        print(sample)
      else:
        sample={}
        continue
      data.append(sample)
      sum+=1
      if sum%20 ==0:
        print('___________目前爬取了 {} 条新闻_______时间 {}'.format(sum,sample['time']))

    if data[-1]['time']<start_time:
        break

  result[keyword]=data

    # t=js['results'][0]['richSnippet']
    # if 'newsarticle' in t:
    #   print('headline:',t['newsarticle']['headline'])
    #   print('url',t['metatags']['url'])
    #   print('time',t['newsarticle']['datepublished'])
    #   print('time2',t['newsarticle']['datemodified'])



#save data
with open('thediplomat.json','w',encoding='utf-8') as f:
  f.write(json.dumps(result,ensure_ascii=False))
```
