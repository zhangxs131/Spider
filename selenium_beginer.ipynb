然后再python代码中进行尝试selenium中的爬虫操作

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
#配置option
options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')


browser= webdriver.Chrome(options=options)
```

下面选择get网页url，进行爬取

```python
#选择google搜索url
browser.get('https://www.google.com')
#选择器，使用NAME，选择查询框
print(browser.title)
input=browser.find_element(by=By.NAME,value='q')
input.send_keys('Deep Learning application in Cyberspace Security')
input.send_keys(Keys.ENTER)
print(browser.title)
```

#selenium被检测识别
现在不少网站采取监测机制，比如淘宝等防止抢单
window.navigator.webdriver 值为undefined，而使用selenium则为true
需要Chromedriver启动参数修改



*   option = ChromeOptions()
*   option.add_experimental_option('excludeSwitches', ['enable-automation'])


```python
browser.get('https://www.taobao.com/')
print(browser.title)

options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
options.add_experimental_option('excludeSwitches', ['enable-automation'])


browser=webdriver.Chrome(options=options)
browser.get('https://www.taobao.com/')
print(browser.title)
```
